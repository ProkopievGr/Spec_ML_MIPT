```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
```

```python
data = pd.read_csv('data.csv')
data
```

```python
data_trunc = data.head(100000)
data_trunc
```

```python
del data
```

```python
X = data_trunc[['latitude', 'longitude']]
```

```python
model = MeanShift(bandwidth=0.1, n_jobs=-1)
model.fit(X)
```

```python
clusters_size = np.unique(model.labels_, return_counts=True)

# indexes of main clusters:
main_clusters = clusters_size[0][np.where(clusters_size[1] > 15)]
```

```python
# for x in model.cluster_centers_[main_clusters]:
#     print(x[0], x[1])
```

```python
offices_coords = np.array([
    [33.751277, -118.188740],
    [25.867736, -80.324116],
    [51.503016, -0.075479],
    [52.378894, 4.885084],
    [39.366487, 117.036146],
    [-33.868457, 151.205134]
])
```

```python
cluster_min_dist = []

for center in model.cluster_centers_[main_clusters]:
    distances_to_offices = np.linalg.norm(center - offices_coords, axis=1)
    
    cluster_min_dist.append(min(distances_to_offices))
```

```python
closest_clusters20 = sorted(zip(main_clusters, cluster_min_dist), key=lambda x: x[1])[:20]

id_closest_cluster = closest_clusters20[0][0]
id_closest_cluster
```

```python
closest_cluster = model.cluster_centers_[id_closest_cluster]

with open('answer.txt', 'w') as f:
    f.write(str(closest_cluster[0]) + ' ' + str(closest_cluster[1]))
```
