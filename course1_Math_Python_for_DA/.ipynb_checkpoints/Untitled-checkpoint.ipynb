```python
import re
```

```python
f = open('sentences.txt')
word_list = set()
for line in f:
    words = [w for w in re.split('[^a-z]', line.lower()) if w != '']
    for w in words:
        word_list.add(w)
        
word_list
```

```python
words_to_index = {w:i for i, w in enumerate(word_list)}
words_to_index
```

```python
a = np.zeros((22, 254))
```

```python
f = open('sentences.txt')
for i, line in enumerate(f):
    words = [w for w in re.split('[^a-z]', line.lower()) if w != '']
    for w in words:
        j = words_to_index[w]
        a[i, j] += 1
```

```python
f = lambda x: np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
```

```python
nodes = np.array([1, 4, 10, 15])
```

```python
A = np.empty((4, 4))
```

```python
for i in range(4):
    A[:, i] = nodes**i
```

```python
A
```

```python
y = f(nodes)
```

```python
coeffs = np.linalg.solve(A, y)
```

```python
sub2 = open('submission-2.txt', 'w')
sub2.write(' '.join(list(map(str, coeffs))))
sub2.close()
```
