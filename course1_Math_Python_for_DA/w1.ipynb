```python
import re
```

```python
f = open('sentences.txt')
word_list = set()
for line in f:
    words = [w for w in re.split('[^a-z]', line.lower()) if w != '']
    for w in words:
        word_list.add(w)
        
word_list
```

```python
words_to_index = {w:i for i, w in enumerate(word_list)}
words_to_index
```
