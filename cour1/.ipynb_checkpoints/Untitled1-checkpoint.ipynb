```python
import numpy as np
import math
```

```python
f = lambda x: math.sin(x/5.)*math.exp(x/10.)+5*math.exp(-x/2.)
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 30)
plt.plot(x, [f(z) for z in x])
plt.show()
```

```python
from scipy.optimize import minimize
```

```python
x1 = np.round(f(minimize(f, 2, method='BFGS').x[0]), 2)
x2 = np.round(f(minimize(f, 30, method='BFGS').x[0]), 2)
```

```python
x1, x2
```

```python
file = open('w31.txt', 'w')
file.write(str(x1) + ' ' + str(x2))
file.close()
```

```python
from scipy.optimize import differential_evolution
```

```python
x3 = differential_evolution(f, [(1, 30)]).x[0]
```

```python
minimize(f, 2, method='BFGS')
```

```python
differential_evolution(f, [(1, 30)])
```

```python
file = open('w32.txt', 'w')
file.write(str(np.round(f(x3), 2)))
file.close()
```

```python
h = lambda x: int(f(x))
```

```python
plt.plot(x, [h(z) for z in x])
plt.show()
```

```python
x4 = minimize(h, 30, method='BFGS').x[0]
x5 = differential_evolution(h, [(1, 30)]).x[0]
with open('w33.txt', 'w') as file:
    file.write(str(h(x4)) + ' ' + str(h(x5)))
```
