```python
import numpy as np
from sklearn.datasets import load_digits, load_breast_cancer
```

```python
def write_answer(ans, i):
    with open('ans' + str(i) + '.txt', 'w') as f:
        f.write(str(ans))
```

```python
X_digits, y_digits = load_digits(return_X_y=True)
X_breast, y_breast = load_breast_cancer(return_X_y=True)
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
```

```python
from sklearn.model_selection import cross_val_score
```

```python
clf1 = BernoulliNB()
clf2 = MultinomialNB()
clf3 = GaussianNB()
```

```python
breast_score = []
for clf in [clf1, clf2, clf3]:
    breast_score.append(cross_val_score(clf, X_breast, y_breast).mean())

write_answer(max(breast_score), 1)
```

```python
digits_score = []
for clf in [clf1, clf2, clf3]:
    digits_score.append(cross_val_score(clf, X_digits, y_digits).mean())

write_answer(max(digits_score), 2)
```
