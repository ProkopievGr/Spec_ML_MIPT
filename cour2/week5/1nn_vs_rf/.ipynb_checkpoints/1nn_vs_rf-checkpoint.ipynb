```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
def write_answer(ans, i):
    with open('ans' + str(i) + '.txt', 'w') as f:
        f.write(str(ans))
```

```python
X, y = load_digits(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=False)
```

```python
def predict_1_nn(x):
    dist = np.diag((X_train - x) @ (X_train - x).T)
    return sorted(zip(dist, y_train), key=lambda x: x[0])[0][1]
```

```python
y_pred = []
for i in range(len(X_test)):
    y_pred.append(predict_1_nn(X_test[i]))
```

```python
err_1 = sum(y_test != y_pred)/len(y_test)
write_answer(err_1, 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=1000)

rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
err_2 = sum(y_pred_rf != y_test) / len(y_test)
```

```python
err_1, err_2
```

```python
write_answer(err_2, 2)
```
