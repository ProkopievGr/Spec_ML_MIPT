```python
import numpy as np
from scipy.stats import norm
```

```python
p = 1/50
alpha = 0.05
z = norm.ppf(1 - alpha/2)
frac = np.sqrt(p * (1 - p) / 50)

print(np.round([p - z * frac, p + z * frac], 4))
```

```python
from statsmodels.stats.proportion import proportion_confint
```

```python
print(np.round(proportion_confint(1, 50, method='normal'), 4))
```

```python
print(np.round(proportion_confint(1, 50, method='wilson'), 4))
```

```python
from statsmodels.stats.proportion import samplesize_confint_proportion
```

```python
np.ceil(samplesize_confint_proportion(1/50, 0.01, alpha=0.05))
```

```python
import matplotlib.pyplot as plt
```

```python
p_range = np.linspace(0, 1, 50)
sample_size_range = [samplesize_confint_proportion(p, 0.01, alpha=0.05) for p in p_range]

plt.figure(figsize=(12, 8))
plt.plot(p_range, sample_size_range)
plt.grid()
# plt.xticks(p_range)
plt.show()
```
