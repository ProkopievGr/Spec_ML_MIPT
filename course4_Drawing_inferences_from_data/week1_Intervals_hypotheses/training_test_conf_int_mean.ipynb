```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('water.txt', sep='\t')
df
```

```python
from statsmodels.stats.weightstats import _tconfint_generic
```

### Вопрос 2
Постройте 95% доверительный интервал для средней годовой смертности в больших городах. Чему равна его нижняя граница? Округлите ответ до 4 знаков после десятичной точки. 

```python
mort_mean = df.mortality.mean()
mort_std = df.mortality.std(ddof=1)/np.sqrt(df.shape[0])

mort_conf_int = _tconfint_generic(mort_mean, mort_std, df.shape[0] - 1, 0.05, 'two-sided')

print(np.round(mort_conf_int, 4))
```

### Вопрос 3
На данных из предыдущего вопроса постройте 95% доверительный интервал для средней годовой смертности по всем южным городам. Чему равна его верхняя граница? Округлите ответ до 4 знаков после десятичной точки.


```python
mort_south_mean = df[df.location == 'South'].mortality.mean()
mort_south_std = df[df.location == 'South'].mortality.std(ddof=1)/np.sqrt(df[df.location == 'South'].shape[0])

mort_south_conf_int = _tconfint_generic(mort_south_mean, mort_south_std, df[df.location == 'South'].shape[0] - 1, 0.05, 'two-sided')

print(np.round(mort_south_conf_int, 4))
```

### Вопрос 4

### Вопрос 5
Пересекаются ли 95% доверительные интервалы для средней жёсткости воды в северных и южных городах?

```python
hardness_north = df[df.location == 'North'].hardness
hardness_south = df[df.location == 'South'].hardness

print(_tconfint_generic(hardness_north.mean(), hardness_north.std(ddof=1)/np.sqrt(df[df.location == 'North'].shape[0]), hardness_north.shape[0] - 1, 0.05, 'two-sided'))
print(_tconfint_generic(hardness_south.mean(), hardness_south.std(ddof=1)/np.sqrt(df[df.location == 'South'].shape[0]), hardness_south.shape[0] - 1, 0.05, 'two-sided'))
```

### Вопрос 6
Вспомним формулу доверительного интервала для среднего нормально распределённой случайной величины с дисперсией $ \sigma^2 $ :
$$ \overline{X_n} \pm z_{1 - \frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}} $$

При $ \sigma=1 $ какой нужен объём выборки, чтобы на уровне доверия 95% оценить среднее с точностью $ \pm0.1 $?

$$ z_{1 - \frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}} = 0.1 $$
$$ n = \left( \frac{\sigma \cdot z_{1 - \frac{\alpha}{2}}}{0.1} \right) ^ 2 $$

```python
from scipy.stats import norm
```

```python
norm.ppf(1 - 0.05/2)
```

```python
np.ceil((norm.ppf(1 - 0.05/2) / 0.1)**2)
```
